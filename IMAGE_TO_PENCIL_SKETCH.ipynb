# ***IMAGE TO PENCIL SKETCH***

*Importing important libraries*

```python
import cv2
import matplotlib.pyplot as plt
```

*Reading the original image using matplotlib*

```python
img=cv2.imread("/content/sweksha image.jpg")
plt.imshow(img)
```

*Original image show using matplotlib*

```python
plt.imshow(img[:,:,::-1])
plt.axis(False)
plt.show()
```

*Converting image to RGB scale*

```python
RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(RGB_img)
plt.axis(False)
plt.show()
```

*Converting RGB image to grey scale*

```python
grey_img=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(grey_img)
plt.axis(False)
plt.show()
```

*Inverting the image*

```python
invert_img=cv2.bitwise_not(grey_img)
plt.imshow(invert_img)
plt.axis(False)
plt.show()
```

*Blur the image to outline it*

```python
blur_img=cv2.GaussianBlur(invert_img, (273,273),0)
plt.imshow(blur_img)
plt.axis(False)
plt.show()
```

*Again inverting the blur image*

```python
invblur_img=cv2.bitwise_not(blur_img)
plt.imshow(invblur_img)
plt.axis(False)
plt.show()
```

*Converting the inverted blur image to sketch*

```python
sketch_img=cv2.divide(grey_img,invblur_img, scale=256.0)
plt.imshow(sketch_img)
plt.axis(False)
plt.show()
```

*Sketch image show using matplotlib*

```python
plt.figure(figsize=(14,8))
plt.subplot(1,2,1)
plt.title('Original image', size=18)
plt.imshow(RGB_img)
plt.axis('off')
plt.subplot(1,2,2)
plt.title('Sketch', size=18)
rgb_sketch=cv2.cvtColor(sketch_img, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_sketch)
plt.axis('off')
plt.show()
```

*Saving the sketch image*

```python
cv2.imwrite('sketch.png', sketch_img)
```

*Creating a function of whole process*

```python
def original_to_sketch(image):
    import cv2
    import matplotlib.pyplot as plt
    img=cv2.imread(image)
    RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    grey_img=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    invert_img=cv2.bitwise_not(grey_img)
    blur_img=cv2.GaussianBlur(invert_img, (99,99),0)
    invblur_img=cv2.bitwise_not(blur_img)
    sketch_img=cv2.divide(grey_img,invblur_img, scale=256.0)
    plt.figure(figsize=(14,8))
    plt.subplot(1,2,1)
    plt.title('Original image', size=18)
    plt.imshow(RGB_img)
    plt.axis('off')
    plt.subplot(1,2,2)
    plt.title('Sketch', size=18)
    rgb_sketch=cv2.cvtColor(sketch_img, cv2.COLOR_BGR2RGB)
    plt.imshow(rgb_sketch)
    plt.axis('off')
    plt.show()
    cv2.imwrite('sketch.png', sketch_img)
```

*The function has been created which takes input of the image path and returns the original image along with sketch image side by side.*

```python
image=input("Enter image path: ")
original_to_sketch(image)
```
